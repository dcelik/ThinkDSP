```python
from __future__ import print_function, division
from IPython.display import Audio

import pandas
import thinkdsp
import thinkplot
import numpy
%matplotlib inline
```

## Question 1:

```python
nrows = 1669
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
wave.plot()
```

```python
spectrum1 = wave.make_spectrum()
thinkplot.preplot(1)
spectrum1.plot_power(low=1, linewidth=1, alpha=0.5)
```
