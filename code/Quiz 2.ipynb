```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%precision 3
%matplotlib inline
```

```python
class TriangleChirp(thinkdsp.Chirp):
    def _evaluate(self,ts,freqs):
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        frac, _ = numpy.modf(phases/thinkdsp.PI2)
        ys = numpy.abs(frac - 0.5)
        ys = thinkdsp.normalize(thinkdsp.unbias(ys), self.amp)
        return ys
    
signal = TriangleChirp(start=1000, end=1500)
wave2 = signal.make_wave(duration=2)
spectrum = wave2.make_spectrum()
spectrum.plot()
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
wave = signal.make_wave(duration=0.5, framerate=11025)
spectrum = wave.make_spectrum()
angle = numpy.angle(spectrum.hs)
f_max = spectrum.framerate / 2.0
fs = numpy.linspace(0, f_max, len(spectrum.hs))
thinkplot.plot(fs, angle)
numpy.mean(angle)
```
