## Question 6:

```python
class SawtoothSignal(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        cycles = self.freq*ts + self.offset / thinkdsp.PI2 #gets the cycles
        frac, _ = numpy.modf(cycles) #get fractional part of cycles, aka the value
        ys = thinkdsp.normalize(thinkdsp.unbias(frac),self.amp) #normalize to mean 0, makes wave go from -1 to 1
        return ys
```

```python
signal = SawtoothSignal(880)
wave = signal.make_wave(duration=1, framerate=10000)
segment = wave.segment(duration=0.005)
segment.plot()
```

```python
signal = thinkdsp.SawtoothSignal(880)
wave = signal.make_wave(duration=1, framerate=10000)
segment = wave.segment(duration=0.005)
segment.plot()
```

The two signals match, so we can assume that our signal is correct. To code the signal, I looked at the triangle signal code and then figured out what the transformation would be to change a triangle signal into a SawtoothSignal

## Question 10:

```python
trsignal = thinkdsp.SinSignal(freq=880, amp=1, offset=0);
for i in range(2,6):
    trsignal += thinkdsp.SinSignal(freq=880*i, amp=1.0/(i**2.0), offset=0)
wave = trsignal.make_wave(duration=1, framerate=10000)
spectrum = wave.make_spectrum()
spectrum.plot()
```

By combining a number of sin signals at both odd and even harmonics with amplitudes that decrease as $1/f^2$, we can construct the desired spectrum.
