We know that for any sinusoid, the time shift is equal to the phase shift divided by the angular frequency:

$Time Shift = \frac{\phi_0}{2\pi f}$

Rearranging the equation we can solve for the necessary phase shift:

$\phi_0 = 2\pi f* Time Shift$

So we multiply each component of the signal to be shifted by :

$e^{j*2*\pi*f_{component}*Time Shift}$

As shown in the graph above, this provides a time shift of 0.005 seconds, half of the period of the complex sinusoid.

# Question 2:

Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
def dft(ys):
    """Function that evaluates a DFT slowly.

    ys: array of amplitudes
    """        
    N = ys.shape[0]#set length of array to N
    n = np.arange(N)#make list n with time values
    k = n.reshape((N,1))#transpose n into k
    M = np.exp(-2j*np.pi*k*n/N)#create fourier transform matrix
    return np.dot(M, ys)# dot matrix with original data to transform
```

```python
sig = thinkdsp.SinSignal()
wave = sig.make_wave()
spec1 = wave.make_spectrum()
spec2 = thinkdsp.Spectrum(hs=np.fft.fft(wave.ys),framerate=wave.framerate)
spec3 = thinkdsp.Spectrum(hs=dft(wave.ys),framerate=wave.framerate)
print(np.allclose(spec2.hs,spec3.hs))
%timeit dft(wave.ys)
%timeit np.fft.fft(wave.ys)
```

## Question 3:

In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.

Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

def cooleytukey(wavearray):
    def wrapper(ys):
        N = ys.shape[0]
        
        if N <= 16:
            return dft(ys)
        evenpart = cooleytukey(ys[::2])
        oddpart = cooleytukey(ys[1::2])
        
        combined = [0] * N
        for m in xrange(N//2):
            combined[m] = evenpart[m] + np.exp(-2j*np.pi*m/N) * oddpart[m]
            combined[m + N//2] = evenpart[m] - np.exp(-2j*np.pi*m/N) * oddpart[m]
        return np.array(combined)
    
    length = wavearray.shape[0]
    w = zero_pad(wavearray,int(2**(np.ceil(np.log2(length)))))
    return wrapper(w)
```

```python
sig3 = thinkdsp.SinSignal()
wave3 = sig3.make_wave()
spec13 = wave3.make_spectrum()
spec23 = thinkdsp.Spectrum(hs=np.fft.fft(zero_pad(wave3.ys,2**(np.ceil(np.log2(len(wave3.ys)))))),framerate=wave3.framerate)
spec33 = thinkdsp.Spectrum(hs=cooleytukey(wave3.ys),framerate=wave3.framerate)
print(np.allclose(spec23.hs,spec33.hs))
%timeit cooleytukey(wave3.ys)
%timeit np.fft.fft(wave3.ys)
```

## Question 4:

Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

##### Defining autocorr from Chapter 5

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** np.arange(6, 14)
print(ns)
```

##### Timing autocorr

```python
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o autocorr(thinkdsp.Wave(ys,framerate = noise.framerate))
    results.append(result)
    
bests = [result.best for result in results]

x = np.log(ns)
y = np.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

##### Defining fft_autocorr from Chapter 8

```python
def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = np.fft.fft(signal)
    fft_window = np.fft.fft(window)
    return np.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

##### Timing fft_autocorr

```python
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o fft_autocorr(ys)
    results.append(result)
    
bests = [result.best for result in results]

x = np.log(ns)
y = np.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

##### Timing numpy.correlate

```python
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o np.correlate(ys,ys)
    results.append(result)
    
bests = [result.best for result in results]

x = np.log(ns)
y = np.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

From the graphs, it is hard to tell if np.correlate uses the fft algorithm, but I would venture to guess that it does since convolving the fft seems to be a relatively quick way. Numpy probably uses lower level methods and less memory overhead to speed it up plus small tricks for certain cases.
