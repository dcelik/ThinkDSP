## Question 2

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(wave)
thinkplot.plot(lags,corrs)
thinkplot.config(xlabel='lag', 
                 ylabel='correlation', 
                 xlim=[-1, 850], 
                 ylim=[-0.05, 1.05])
```

The autocorrelation function does not seem to decay exponentially. There is no evidence of periodic behavior.

## Question 4

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze2(ys, freqs, ts)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o scipy.fftpack.dct(ys)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

analyze2 seems to be about double the speed of analyze1, but scipy.fftpack.dct is significantly faster as all of its best were in the microseconds, while the two analyze functions were sometimes in the milliseconds. The graphs also corroborate this fact as the analyze 2 graph has overall lower values, while the scipy graph has little gain in time as the number of loops increases.

## Question 5

##### 1. Break a long signal into segments:

```python
wave5 = thinkdsp.read_wave('NunuBot.wav')
wave5.make_audio()
```

```python
wave5.plot()
```

```python
n = 50
dur = wave5.duration/n
frmrt=wave5.framerate

segments = []
for i in range(0,n):
    segi = wave5.segment(start=i/(n/wave5.duration),duration=dur)
    segments.append(segi)
```

##### 2. Compute DCT of each segment

```python
dcts = []
for s in segments:
    ys = s.ys.astype(dtype='float')
    dcts.append(scipy.fftpack.dct(ys))
```

##### 3. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.

```python
savedamps = []
saveddcts = []
for d in dcts:
    savedamps = []
    for j in d:
        if abs(j)>3000000:
            savedamps.append(j)
        else:
            savedamps.append(0.0)
    saveddcts.append(savedamps)
```

##### 4. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.

```python
freqs = []
for sd in saveddcts:
    if(len(sd) is not 0):
        freqs.append(scipy.fftpack.idct(sd))#

finalwave = thinkdsp.Wave([],frmrt)       
for f in freqs:
        w = thinkdsp.Wave(f,frmrt)
        finalwave = finalwave | w

finalwave.make_audio()
    
```

Now we analyze the original wave and spectrum versus the DCT version.

```python
wave5.plot()
```

Original Wave

```python
finalwave.plot()
```

DCT Wave

```python
spec = wave5.make_spectrum()
spec.plot(high = 20000)
```

Original Wave Spectrum

```python
spec2 = finalwave.make_spectrum()
spec2.plot(high = 20000)
```

DCT Wave Spectrum

It looks like all of the frequencies with lower amplitudes were removed and all that remains are the frequencies with the largest amplitudes. The modified wave is similar enough to be identified as the same source, but the quality is terrible in comparison.
